```python
import urllib
import requests
import json
import csv
import pandas as pd
```

```python
url = "https://covid.ourworldindata.org/data/owid-covid-data.json"
```

```python
f = requests.Session()
```

```python
response = f.get(url,stream = True)
```

```python
data = response.json()
```

```python
data.keys()
```

### Continent list

```python
continent =[]
for i in data:
    str = data[i].get('continent')
    continent.append(str)
print(continent)
```

### country code

```python
Country_Code = []
#count = 0
for ele in data:
    #if(count == 40):
     #   break
    #else:
    Country_Code.append(ele)
    #count = count+1
print(Country_Code)
```

```python
data['IND'].get('data')
```

```python
final_df = pd.DataFrame(columns=['Country','Country_Code', 'Date','Total_Cases','Total_Deaths','New_Deaths','New_Cases',
                                 'Total_Tests','Tests_Units','Total_Vaccinations','New_Vaccinations'])
final_df
```

```python
for cnty in Country_Code:
    tmp_dict = {'Country':[], 'Country_Code':[], 'Date':[],'Total_Cases': [],'Total_Deaths':[], 'New_Deaths':[],
                'New_Cases':[],'Total_Tests':[],'Tests_Units':[],'Total_Vaccinations':[],
                'New_Vaccinations':[]}
    
    loc = data[cnty].get('location')
    code = cnty
    for curr_dict in data[cnty].get('data'):
        tmp_date = curr_dict.get('date')
        tmp_cases = curr_dict.get('total_cases')
        tmp_total_deaths = curr_dict.get('total_deaths')
        tmp_new_deaths = curr_dict.get('new_deaths')
        tmp_new_cases = curr_dict.get('new_cases')
        tmp_total_tests = curr_dict.get('total_tests')
        tmp_test_units = curr_dict.get('tests_units')
        tmp_total_vacc = curr_dict.get('total_vaccinations')
        tmp_new_vacc = curr_dict.get('new_vaccinations')
        
        
        tmp_dict['Country'].append(loc)
        tmp_dict['Country_Code'].append(code)
        tmp_dict['Date'].append(tmp_date)
        tmp_dict['Total_Cases'].append(tmp_cases)
        tmp_dict['Total_Deaths'].append(tmp_total_deaths)
        tmp_dict['New_Deaths'].append(tmp_new_deaths)
        tmp_dict['New_Cases'].append(tmp_new_cases)
        tmp_dict['Total_Tests'].append(tmp_total_tests)
        tmp_dict['Tests_Units'].append(tmp_test_units)
        tmp_dict['Total_Vaccinations'].append(tmp_total_vacc)
        tmp_dict['New_Vaccinations'].append(tmp_new_vacc)
        
    final_df = final_df.append(pd.DataFrame.from_dict(tmp_dict), ignore_index=True)
```
